```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.cluster import SpectralClustering
from unsupervised_helper_functions import *
```

```python
# Importamos los datos
df = pd.read_csv('data/spirals.csv')
df.columns = ['x', 'y']
df.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(df['x'], df['y'], c='#E65983', edgecolors='black', s=50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Spirals', loc='left', fontdict={'fontsize': 15, 'fontweight': 'bold'})
```

```python
spectral = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=3, random_state=0).fit(df)
y_pred = spectral.labels_
```

```python
plot_clusters(data=df, labels=y_pred, title='Spectral Clustering')
```
