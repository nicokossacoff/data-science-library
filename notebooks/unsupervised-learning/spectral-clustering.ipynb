```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.cluster import SpectralClustering
from unsupervised_helper_functions import *
```

```python
# Importamos los datos
df = pd.read_csv('data/spirals.csv')
df.columns = ['x', 'y']
df.head()

# Visualizamos los datos
plot_data(df, title='Spirals dataset')
```

Ajustamos el modelo. La matriz de similaridad la construimos utilizando la estrategia de kNN.

```python
spectral = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=3, random_state=0).fit(df)
y_pred = spectral.labels_
```

```python
plot_clusters(data=df, labels=y_pred, title='Spectral Clustering')
```

```python
estimator = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=3, random_state=0)
hening_stability(X=df, estimator=estimator, y_true=y_pred, K=2, B=1000, random_state=0)
```
